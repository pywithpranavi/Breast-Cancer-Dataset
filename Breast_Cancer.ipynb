```python
x=df.iloc[:,0:-1].values
y=df.iloc[:,-1].values
```

```python
print("shape of x:",x.shape)
print("shape of y:",y.shape)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=51)
```

```python
print("shape of xtrain:",xtrain.shape)
print("shape of ytrain:",ytrain.shape)
print("shape of xtest:",xtest.shape)
print("shape of ytest:",ytest.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rm_g=RandomForestClassifier(n_estimators=100,criterion='gini')
rm_g.fit(xtrain,ytrain)
rm_g.score(xtest,ytest)
```

```python
rm_e=RandomForestClassifier(n_estimators=120,criterion='entropy')
rm_e.fit(xtrain,ytrain)
rm_e.score(xtest,ytest)
```

```python
rm_e=RandomForestClassifier(n_estimators=100,criterion='entropy')
rm_e.fit(xtrain,ytrain)
rm_e.score(xtest,ytest)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,rm_e.predict(xtest))
```

```python
ypred=rm_e.predict(xtest)
```

```python
cm=confusion_matrix(ytest,ypred)
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```
